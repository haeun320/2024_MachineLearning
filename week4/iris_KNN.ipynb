```python
import seaborn as sns
iris = sns.load_dataset('iris')
print(iris.head())
```

```python
print(iris.shape)
```

X: 입력 데이터
y: 라벨 타겟

```python
X = iris.drop('species', axis = 1)
print(X.shape)
```

```python
y = iris['species']
```

카테고리의 실수화

```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
classle = LabelEncoder()
y = classle.fit_transform(iris['species'].values)
print('species labels:', np.unique(y))
```

```python
#원래의 문자열로 전환
yo = classle.inverse_transform(y)
print('species:', np.unique(yo))
```

데이터 분할

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

모형 추정 및 사례 중심 학습

```python
#KNN 학습
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, p=2) #5개의 인접한 이웃, 거리 측정 기준: 유클리드

#모델 fitting -> 학습
knn.fit(X_train, y_train)
```

```python
#예측, 평가
y_train_pred = knn.predict(X_train) #train data의 y값 예측치
y_test_pred = knn.predict(X_test) #모델을 적용한 test data의 y값 예측치

print('Misclassified training samples: %d' %(y_train != y_train_pred).sum()) #오분류 데이터 갯수 확인
print('Misclassified test samples: %d' %(y_test != y_test_pred).sum())
```

결과 분석

```python
#정확도 계산
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_test_pred))
```
