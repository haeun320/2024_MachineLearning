카테고리의 실수화

```python
from sklearn.preprocessing import LabelEncoder #LabelEncoder() method를 불러옴
import numpy as np
```

```python
classle = LabelEncoder()
y = classle.fit_transform(iris['species'].values) #species의 문자열 숫자로 자동 매핑
```

```python
print('species labels:', np.unique(y)) #unique():중복제거
```

```python
yo = classle.inverse_transform(y)
print('species:', np.unique(yo))
```
